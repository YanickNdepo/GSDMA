```python
import pyvista as pv
from pyvista import examples
```

```python
# load a sample texture
textures = [examples.download_masonry_texture(), examples.download_puppy_texture()]
colors = ['red', 'green', 'blue']
# create a surface to host this texture
intervals = {}
for f in range(200):
    intervals.update({f:{'shape':pv.Cylinder(center=(0, 0, f*1.)), 'texture': textures[f%len(textures)], 'color': colors[f%len(colors)]}})
```

```python
intervals
```

```python
pl = pv.Plotter()
for k, v in intervals.items():
    pl.add_mesh(v['shape'], texture=v['texture'])
pl.show() 
```

```python
pl = pv.Plotter()
for k, v in intervals.items():
    pl.add_mesh(v['shape'], color=v['color'])
pl.show() 
```
