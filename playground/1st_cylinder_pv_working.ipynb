```python
import pyvista as pv
import pandas as pd
import numpy as np
```

```python
cyl = pv.Cylinder(
    center=(0.0, 0.0, 0.0),
    direction=(0.0, 0.0, 1.0),
    radius=0.1,
    height=5.0,
    resolution=10,
    
)
#?pv.Cylinder 
cyl
```

```python
cyl_structured = pv.CylinderStructured(
    radius=0.5, height=10.0, center=(0.0, 0.0, 0.0), direction=(1.0, 0.0, 0.0), theta_resolution=32, z_resolution=10)
```

```python
p = pv.Plotter(shape=(1,1))
pp = pv.Plotter(shape=(1,1))
```

```python
? pv.Plotter
```

```python
p.add_mesh(cyl, color="tan", show_edges=True)
p.show()
```

```python
pp.add_mesh(cyl_structured, color="tan", show_edges=True)
pp.show()
```
